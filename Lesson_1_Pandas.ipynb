```python
import pandas as pd
```

```python
authors = pd.DataFrame({'author_id':[1, 2, 3], 'author_name':['Тургенев', 'Чехов', 'Островский']}, columns = ['author_id', 'author_name'])
authors
```

```python
book = pd.DataFrame({'author_id':[1, 1, 1, 2, 2, 3, 3], 'book_title':['Отцы и дети', 'Рудин', 'Дворянское гнездо', 'Толстый и тонкий', 'Дама с собачкой', 'Гроза', 'Таланты и поклонники'], 'price':[450, 300, 350, 500, 450, 370, 290]}, columns = ['author_id', 'book_title', 'price'])
book
```

```python
authors_price = pd.merge(book, authors, on = 'author_id', how = 'outer')
authors_price
```

```python
top5 = authors_price.nlargest(5, 'price')
top5
```

```python
authors_stat = authors_price.groupby('author_name')['price'].agg(['max', 'min', 'mean'])
authors_stat.columns = ['max_price', 'min_price', 'mean_price']
authors_stat
```

```python
authors_price['cover'] = ['твердая', 'мягкая', 'мягкая', 'твердая', 'твердая', 'мягкая', 'мягкая']
authors_price
```

```python
import numpy as np
```

```python
book_info = pd.pivot_table(authors_price, index = 'author_name', columns = 'cover', values = 'price', aggfunc = np.sum, fill_value = 0)
book_info
```

```python
book_info.to_pickle('book_info.pkl')
```
